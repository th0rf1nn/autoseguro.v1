```python
#tagucci
import random

lista_vehiculo = []




def mostrar_menu():
    print('''
         __________________________
        |=======AUTO SEGURO========|
        |__________________________|
        |__________________________|
        |     ¡Menú Principal!     |
        |__________________________|
        |  1. Grabar vehículo      |
        |__________________________|
        |  2. Buscar vehículo      |
        |__________________________|
        |  3. Imprimir certificados|
        |__________________________|
        |  4. Salir                |
        |__________________________|
        ''')



def grabar_vehiculo():
    tipo_vehiculo = input("Ingrese el tipo de vehículo: ")
    patente_vehiculo = input("Ingrese la patente del vehículo: ")

    while len(patente_vehiculo) != 6:
        print("La patente debe tener 6 caracteres.")
        patente_vehiculo = input("Ingrese la patente del vehículo: ")

    respuesta = input("¿Tiene multas? (si/no): ")

    if respuesta.lower() == "si":
        lista_multas = []
        while True:
            fecha_multa = input("Ingrese la fecha de la multa (o 'salir' para finalizar): ")
            if fecha_multa.lower() == "salir":
                break
            valor_multa = float(input("Ingrese el valor de la multa: "))
            lista_multas.append({"Fecha": fecha_multa, "Valor": valor_multa})
    else:
        lista_multas = []

    marca_vehiculo = input("Ingrese la marca del vehículo: ")
    valor_vehiculo = float(input("Ingrese el precio del vehículo: "))
    fecha_registro = input("Ingrese la fecha de registro del vehículo: ")
    nombre_dueno = input("Ingrese el nombre del dueño del vehículo: ")

    if len(marca_vehiculo) < 2 or len(marca_vehiculo) > 15:
        print("La marca debe tener entre 2 y 15 caracteres.")
        return

    if valor_vehiculo <= 5000000:
        print("El precio debe ser mayor a $5.000.000.")
        return

    inf_vehiculo = {
        "Tipo": tipo_vehiculo,
        "Patente": patente_vehiculo,
        "Marca": marca_vehiculo,
        "Precio": valor_vehiculo,
        "Multas": lista_multas,
        "Fecha de Registro": fecha_registro,
        "Nombre del Dueño": nombre_dueno
    }

    lista_vehiculo.append(inf_vehiculo)
    print("Vehículo grabado exitosamente.")


def buscar_vehiculo():
    vehiculo_encontrado = input("Ingrese la patente del vehículo a buscar: ")

    for vehiculo in lista_vehiculo:
        if vehiculo["Patente"] == vehiculo_encontrado:
            print("|El vehiculo ha sido encontrado")
            print("|Nombre del Dueño:", vehiculo["Nombre del Dueño"])
            print("|Marca:", vehiculo["Marca"])
            print("|Precio:", vehiculo["Precio"])
            print("|Multas:", vehiculo["Multas"])
            print("|Fecha de Registro:", vehiculo["Fecha de Registro"])
            print("|Patente:", vehiculo["Patente"])
            break

    else:
        print("Vehículo no encontrado.")


def imprimir_certificado():
    for vehiculo in lista_vehiculo:
        patente = vehiculo["Patente"]
        dueno = vehiculo["Nombre del Dueño"]
        emisiones_contaminantes = random.randint(1500, 3500)
        anotaciones_vigentes = random.randint(1500, 3500)
        multas = valor_multa

        print("|Certificado de emisiones contaminantes")
        print("|Patente:", patente)
        print("|Dueño:", dueno)
        print("|Valor: $", emisiones_contaminantes)
        print()

        print("|Certificado de anotaciones vigentes")
        print("|Patente:", patente)
        print("|Dueño:", dueno)
        print("|Valor: $", anotaciones_vigentes)
        print()

        print("|Certificado de multas")
        print("|Patente:", patente)
        print("|Dueño:", dueno)
        print("|Valor: $", multas)
        print()


while True:
    mostrar_menu()
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        grabar_vehiculo()
    elif opcion == "2":
        buscar_vehiculo()
    elif opcion == "3":
        imprimir_certificado()
    elif opcion == "4":
        print('''Gracias por usar el programa!
        Lucas Hernandez
        Nicolás Ponce
        AutoSeguro.v1
        ''')
        break
```
